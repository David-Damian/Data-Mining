```python
fig = plt.figure(1, figsize=(35, 20))
gs = gridspec.GridSpec(2,4)
gs.update(wspace=.15, hspace=0.25)


#Generate first panel
#remember, the grid spec is rows, then columns
xtr_subsplot= fig.add_subplot(gs[0:1,0:1])


x=datos_train['Age']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 10

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.legend(loc='upper right')
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Edades', size=15, fontweight="bold")
plt.ylabel('Porcentaje de la poblacion', size=20,fontweight="bold")


xtr_subsplot= fig.add_subplot(gs[0:1,1:2])

x=datos_train['Annual_Income']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 10

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Ingresos Anuales', size=15, fontweight="bold")


xtr_subsplot= fig.add_subplot(gs[0:1,2:3])

x=datos_train['Monthly_Inhand_Salary']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 10

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Ingresos Mensuales', size=15, fontweight="bold")


xtr_subsplot= fig.add_subplot(gs[0:1,3:4])

x=datos_train['Num_Bank_Accounts']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 15

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Cuentas Bancarias', size=15, fontweight="bold")


xtr_subsplot= fig.add_subplot(gs[1:2,0:1])

x=datos_train['Num_Credit_Card']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 15

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Numero de Tarjetas de Credito', size=15, fontweight="bold")
plt.xlim([0,15])
plt.ylabel('Porcentaje de la poblacion', size=20,fontweight="bold")

xtr_subsplot= fig.add_subplot(gs[1:2,1:2])

x=datos_train['Interest_Rate']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 15

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Tasa de Interes', size=15, fontweight="bold")
plt.ylabel('Porcentaje de la poblacion', size=20,fontweight="bold")
plt.xlim([0,45])


xtr_subsplot= fig.add_subplot(gs[1:2,2:3])

x=datos_train['Num_of_Loan']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 15

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Numero de Creditos', size=15, fontweight="bold")
plt.xlim([0,12])


xtr_subsplot= fig.add_subplot(gs[1:2,3:4])

x=datos_train['Delay_from_due_date']
x_w = np.empty(x.shape)
x_w.fill(100/x.shape[0])
bins = 15

plt.hist(x, bins, weights=[x_w],color="blue",alpha=0.6)
plt.grid(axis="y",alpha=0.7, linestyle="dotted")
plt.title('Dias de retraso en\n el pago de\n creditos', size=15, fontweight="bold")

plt.savefig('/home/juanpalms/Desktop/Mineria de Datos/Proyecto/Graficas_resultados/histograma1_explicativas.png', dpi=300,bbox_inches="tight")
```
